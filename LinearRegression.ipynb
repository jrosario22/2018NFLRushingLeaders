```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf
import numpy as np

%matplotlib inline
```

```python
rushing = pd.read_csv("2018_NFL_Rushing_Stats.csv")
```

```python
rushing
```

```python
lm = smf.ols('TD ~ Yds', rushing).fit()
lm.summary()
```

```python
sns.regplot(y = "TD", x = "Yds", data = rushing, fit_reg = True)
plt.title("Linear Model")
```

```python
lm.resid.hist(bins = 20)
plt.xlabel("Residual")
plt.ylabel("Frequency")
plt.title("Residuals of linear model")
```

```python
plt.scatter(rushing['TD'], lm.fittedvalues)
plt.xlabel("Touchdowns")
plt.ylabel("Predicted Touchdowns")
plt.title("Touchdowns vs Predicted Touchdowns")
```
